```python
#!/usr/bin/env python
# Make sure to run:
# !apt-get update && apt-get install ffmpeg -y  # Install ffmpeg

import os
import subprocess
import re
import json
import whisper
from transformers import pipeline

def extract_audio(video_path, audio_path="output_audio.wav"):
    """Extracts audio from video using ffmpeg."""
    command = [
        "ffmpeg",
        "-i", video_path,      # Input video file
        "-ac", "1",            # Convert to mono audio
        "-ar", "16000",        # Set sample rate to 16kHz
        "-y",                  # Overwrite output file if exists
        audio_path             # Output audio file
    ]
    subprocess.run(command, check=True)
    return audio_path

def transcribe_audio(audio_path, model_size="medium"):
    """Transcribes audio using OpenAI Whisper."""
    model = whisper.load_model(model_size)  # Load the Whisper model
    result = model.transcribe(audio_path)     # Transcribe audio
    return result

def save_transcript(transcript, output_path="transcript.txt"):
    """Saves transcript as a text file."""
    with open(output_path, "w", encoding="utf-8") as f:
        f.write(transcript)
    print(f"âœ… Transcript saved as: {output_path}")

def save_srt(transcript_result, output_path="subtitles.srt"):
    """Saves transcript as an SRT file with timestamps."""
    segments = transcript_result.get("segments", [])
    with open(output_path, "w", encoding="utf-8") as f:
        for i, segment in enumerate(segments):
            start_time = segment["start"]
            end_time = segment["end"]
            text = segment["text"].strip()
            f.write(f"{i+1}\n")
            f.write(f"{format_srt_time(start_time)} --> {format_srt_time(end_time)}\n")
            f.write(f"{text}\n\n")
    print(f"âœ… Subtitles saved as: {output_path}")

def format_srt_time(seconds):
    """Converts seconds to SRT timestamp format (HH:MM:SS,mmm)."""
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    secs = int(seconds % 60)
    milliseconds = int(round((seconds - int(seconds)) * 1000))
    return f"{hours:02}:{minutes:02}:{secs:02},{milliseconds:03}"

def search_transcript(transcription_result, keyword):
    """
    Searches transcript segments for the given keyword and prints matching segments
    with timestamps. The keyword is highlighted using ANSI escape codes.
    """
    matches = []
    for segment in transcription_result.get("segments", []):
        if keyword.lower() in segment["text"].lower():
            matches.append(segment)
    if not matches:
        print(f"No matches found for keyword: '{keyword}'")
    else:
        print(f"\nğŸ” Found {len(matches)} match(es) for keyword: '{keyword}'")
        for match in matches:
            start = format_srt_time(match["start"])
            end = format_srt_time(match["end"])
            text = match["text"]
            # Highlight the keyword (yellow color)
            highlighted = re.sub(f"({keyword})", r"\033[93m\1\033[0m", text, flags=re.IGNORECASE)
            print(f"[{start} --> {end}] {highlighted}")

def summarize_transcript(transcript, model_name="facebook/bart-large-cnn", max_length=150, min_length=40):
    """Generates a summary of the transcript using a Hugging Face summarization pipeline."""
    summarizer = pipeline("summarization", model=model_name)
    summary = summarizer(transcript, max_length=max_length, min_length=min_length, do_sample=False)
    return summary[0]['summary_text']

def export_json(transcription_result, summary, output_path="output.json"):
    """Exports transcript, subtitle segments, and summary to a JSON file."""
    data = {
         "transcript": transcription_result["text"],
         "segments": transcription_result.get("segments", []),
         "summary": summary
    }
    with open(output_path, "w", encoding="utf-8") as f:
         json.dump(data, f, indent=2)
    print(f"âœ… JSON export saved as: {output_path}")

if __name__ == "__main__":
    # Pass the video file path as a variable:
    video_file = "/content/sample_data/videoplayback.mp4"  # Change this to your video file path
    audio_file = "output_audio.wav"

    # Set your search keyword (e.g., a character or term you want to locate)
    search_keyword = "character"

    print("ğŸ¬ Extracting audio...")
    extract_audio(video_file, audio_file)

    print("ğŸ“ Transcribing audio...")
    transcription_result = transcribe_audio(audio_file, model_size="medium")

    print("ğŸ’¾ Saving transcript...")
    save_transcript(transcription_result["text"], "transcript.txt")

    print("ğŸï¸ Generating subtitles...")
    save_srt(transcription_result, "subtitles.srt")

    if search_keyword:
        search_transcript(transcription_result, search_keyword)

    print("ğŸ“ Summarizing transcript...")
    summary = summarize_transcript(transcription_result["text"])
    print("ğŸ” Summary:", summary)

    print("ğŸ’¾ Exporting JSON...")
    export_json(transcription_result, summary, output_path="output.json")

    # Cleanup: Remove temporary audio file
    os.remove(audio_file)
    print("ğŸ—‘ï¸ Temporary audio file deleted.")
```
